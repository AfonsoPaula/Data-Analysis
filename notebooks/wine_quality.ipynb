```python
# Libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import dataset

```python
df = pd.read_csv("winequality-red.csv")
df
```

## Calculate basic descriptive statistics for numerical variables

```python
df.describe()
```

## Calculate size of a dataset (number of observations and variables)

```python
print("Number of observations: ", df.shape[0])
print("Number of variables: ", df.shape[1])
```

## List structures of the dataset

```python
df.info()
```

## Calculate number of missing values

```python
missing_counts = df.isna().sum()
total_missing = missing_counts.sum()
print("Number of missing values: ", total_missing)
```

```python
print("Number of missing values for each variable: \n")

missing_counts
```

## Box and whisker for each variable

```python
variables = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

df[variables].plot(kind='box', subplots=True, layout=(3,4), figsize=(15,10))
plt.suptitle('Box Plot for each variable', fontsize=16)
plt.show()
```

## Histograms for each variable

```python
# Style for sea born
sns.set(style="whitegrid")

fig, axes= plt.subplots(nrows=3, ncols=4, figsize=(15, 10))

# Flatten the axes array
axes = axes.flatten()

# List of neumeric column
numeric_columns = df.select_dtypes(include=['float64']).columns

# Create histograms
for i, column in enumerate(numeric_columns):
    sns.histplot(df[column], ax=axes[i], kde=True, bins=20, color='blue')
    axes[i].set_title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# For this case we had to apply the 'n' root rule for the "quality" variable (discrete data)
num_bins = int(np.sqrt(len(df['quality'])))

plt.hist(df['quality'], bins=num_bins)
plt.title('Histogram for the quality of the wines', fontsize=16)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()

print(f"\n{np.median(df['quality'])} is the median for -- quality.")
```

## Wine quality

```python
df['quality'].value_counts().plot(kind='pie', figsize=(7,7))
plt.title('Wine Quality (0-10)', fontsize=16)

df['quality'].value_counts()
```

```python
variables = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

# Plot boxplots for each wine component
for column in variables:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='quality', y=column, data=df)
    plt.title(f'Boxplot: {column} - Wine Quality')
    plt.xlabel('')
    plt.ylabel(column)
    plt.show()
    print("\n\n\n\n")
```

```python
variables = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
             'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

# Dictionary
quartile_per_variavel = {}

df_quality_8 = df[df['quality'] == 8]

# Get 1st quartile and 3rd quartil for each component
for column in variables:
    quartiles = df_quality_8[column].quantile([0.25, 0.75])
    quartile_per_variavel[column] = f'\n1 quartile: {quartiles.iloc[0]}, 3 quartile: {quartiles.iloc[1]}'

# Print dictionary
for key, value in quartile_per_variavel.items():
    print(f'{key}: {value}\n')
```

## Scatter Plots

```python
# Calculate Pearson correlation matrix
correlation_matrix = df.corr(method='pearson')

# Filter the correlations for quality 8
correlation_quality_8 = correlation_matrix['quality']

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation Heatmap')
plt.show()

# Identify components with highest correlation with quality 8
highest_correlation_quality_8 = correlation_quality_8[abs(correlation_quality_8) > 0.5]
print("Components with highest correlation with quality 8:")
print(highest_correlation_quality_8)
```

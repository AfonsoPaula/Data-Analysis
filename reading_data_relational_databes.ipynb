## Read data from SQL database

Reading data from SQL relational databases is fairly simple and pandas support a variety of methods to deal with it.

We'll start with an example using SQLite, as it's a builtin Pyhton package, and we don't need anything extra installed.

```python
import sqlite3
```

In order to work with SQLite database from Python, we first have to connect to it. We can do that using the connect function, which returns a **Connection** object.

```python
conn = sqlite3.connect('chinook.db')
```

Once we have a **Connection** object, we can then create a **Cursor** object. Cursors allow us to execute SQL queries against a databese:

```python
cur = conn.cursor()
```

The **Cursor** created has a method **execute**, which will receive SQL parameters to run against the database.

The code below will fetch the first **5** rows from the **employees** table:

```python
cur.execute('SELECT * FROM employees LIMIT 5;')
```

We didn't assign the result of the above query to variable. This is because we need to run another command to actually fetch the results.

We can use the **fetchAll** method to fetch all the results of a query:

```python
results = cur.fetchall()
```

```python
results
```

As we can see, the results are returned as a list of tuples. Each tuple corresponds to a row in the databse that we accessed. Dealing with data this way is painful.

We'd need to manually add column headers, and manually parse the data. Luckily, the pandas library has an easier way, which we'll look at in the next section.

```python
df = pd.DataFrame(results)

df.head()
```

Before we move on, it's good pratice to close **Connection** objects and **Cursor** objects that are open. This prevents the SQLite database from being locked.When a SQLite database is locked, we may be unable to update the database, and may get errors. We can close the Cursor and the Connection like this:

```python
cur.close()
conn.close()
```

## Using pandas **read_sql** method

we can use the pandas **read_sql** function to read the results of a SQL query directly into a pandas **DataFrame**. The code below will execute the same query that we just did, but will return a **DataFrame**. It has several advantages over the query we did above.
- It doesn't require us to create a **Cursor** object or call **fetchall** at the end.
- It automatically reads in the names of the headers from the table.
- It creates a **DataFrame**, so we can quickly explore the data.

```python
conn = sqlite3.connect('chinook.db')
```

```python
df = pd.read_sql('SELECT * FROM employees;', conn)
```

```python
df.head()
```

```python
df = pd.read_sql('SELECT * FROM employees;', conn,
                 index_col='EmployeeId',
                 parse_dates=['BirthDate', 'HireDate'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['ReportsTo'].isna().sum()
```

```python
df['ReportsTo'].mean()
```

```python
df['ReportsTo'] > 1.75
```

```python
df['City'] = df['City'].astype('category')
```

```python
df.info()
```

## Using pandas **read_sql_query** method

It turns out that the **read_sql** method we saw above it just a wrapper around **read_sql_query** and **read_sql_table**.

We can get the same result using **read_sql_query** method:

```python
conn = sqlite3.connect('chinook.db')
```

```python
df = pd.read_sql_query('SELECT * FROM employees LIMIT 5;', conn)
```

```python
df.head()
```

```python
df = pd.read_sql_query('SELECT * FROM employees;', conn,
                       index_col='EmployeeId',
                       parse_dates=['BirthDate', 'HireDate'])

df.head()
```

## Using **read_sql_table** method

**read_sql_table** is a useful function, but it works only with SQLAlchemy, a Python SQL Toolkit and Object Relational Mapper.

This is just a demonstration of its usage where we read the whole **employees** table.

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///chinook.db')

connection = engine.connect()
```

```python
df = pd.read_sql_table('employees', con=connection)
```

```python
df.head()
```

```python
df = pd.read_sql_table('employees', con=connection,
                       index_col='EmployeeId',
                       parse_dates=['BirthDate', 'HireDate'])

df.head()
```

```python
connection.close()
```

## Create tables from **DataFrame** objects

Finally we can persist **DataFrame** objects we've working on in database using the pandas **to_sql** method.

Although it is easy to implement, it could be a very slow process.

```python
df.head()
```

```python
df.to_sql('employees_save', conn)
```

```python
pd.read_sql_query('SELECT * FROM employees_save;', conn).head()
```

```python
#pd.read_sql_query('DROP TABLE employees_save;', conn)
```

## Custom behaviour

The **if_exists** parameter define how to behave if the table already exists and adds a ton of flexibility, letting you decide weather to **replace** current database data, **append** new data at the end, or simply **fail** if database already exists.

```python
pd.DataFrame().to_sql('employees_save',
                      conn,
                      if_exists='replace')
```

```python
pd.read_sql_query('SELECT * FROM employees_save;', conn).head()
```

```python
df.to_sql('employees_save',
          conn,
          if_exists='replace')
```

```python
pd.read_sql_query('SELECT * FROM employees_save;', conn).head()
```

```python
conn.close()
```
